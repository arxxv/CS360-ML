```python
import pandas as pd
from PIL import Image
import requests #to get an image from a link
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1
# getting the image using requests and loading it using PIL 
url = "https://humornama.com/wp-content/uploads/2020/08/Leonardo-DiCaprio-laughing-meme-template-of-Django-Unchained.jpg"
req = requests.get(url, stream=True).raw
img = Image.open(req)
plt.imshow(img)

# getting numpy array
imgnparr = np.asarray(img)
print(imgnparr.shape)

# saving it to a csv
imgdf = pd.DataFrame(imgnparr.reshape(-1,3))
imgdf.to_csv(os.path.join(path, "imgdf.csv"))
```

```python
# 2
# reading it from the csv
imgdf_load = pd.read_csv(os.path.join(path, "imgdf.csv"), header=0, names=['pixel', 'r', 'g'], skipfooter=1, usecols=range(3))
print(imgdf_load.shape)
imgdf_load.head(10)
```
